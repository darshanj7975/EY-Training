```python
# =========================
# Exploratory Data Analysis
# Geyser Dataset
# =========================

# Import Libraries
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# =========================
# Section 1 : Load Dataset
# =========================

df = sns.load_dataset('geyser')

print("Shape of Dataset:")
print(df.shape)

print("\nColumns in Dataset:")
print(df.columns)

print("\nDataset Description:")
print(df.describe())

print("\nFirst 5 Rows:")
print(df.head())

print("\nDataset Information:")
print(df.info())


# =========================
# Section 2 : Data Cleaning
# =========================

# Remove missing values
df_clean = df.dropna()

print("\nCleaned Dataset:")
print(df_clean.head())

print("\nShape After Cleaning:")
print(df_clean.shape)

print("\nCleaned Dataset Info:")
print(df_clean.info())

# Find duplicate values
print("\nDuplicate Values:")
print(df.duplicated().sum())

# Find missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Final cleaned dataset copy
df_clean = df.copy()

print("\nFinal Clean Dataset:")
print(df_clean.head())


# =========================
# Section 3 : Statistical Analysis
# =========================

# Mean
print("\nMean Values:")
print(df_clean.mean(numeric_only=True))

# Median
print("\nMedian Values:")
print(df_clean.median(numeric_only=True))

# Standard Deviation
print("\nStandard Deviation:")
print(df_clean.std(numeric_only=True))

# GroupBy Analysis
res = df.groupby("kind")[["waiting", "duration"]].mean()

print("\nGroupBy Mean Results:")
print(res)

# Pivot Table
pivot = pd.pivot_table(
    df,
    values=["waiting", "duration"],
    index="kind",
    aggfunc="mean"
)

print("\nPivot Table:")
print(pivot)


# =========================
# Section 4 : Correlation and Visualization
# =========================

# Correlation
print("\nCorrelation Matrix:")
print(df_clean.corr(numeric_only=True))

# Group Mean
group_mean = df.groupby('kind')['waiting'].mean()

print("\nGroup Mean:")
print(group_mean)


# Scatter Plot
plt.figure(figsize=(6, 4))

sns.scatterplot(
    data=df_clean,
    x="waiting",
    y="duration"
)

plt.title("Waiting vs Eruption Duration")
plt.xlabel("Waiting Time")
plt.ylabel("Duration")

plt.show()


# Histogram
plt.figure(figsize=(6, 4))

plt.hist(df_clean["waiting"], bins=20)

plt.title("Waiting Time Distribution")
plt.xlabel("Waiting Time")
plt.ylabel("Frequency")

plt.show()


# Box Plot
plt.figure(figsize=(5, 4))

sns.boxplot(y=df_clean["duration"])

plt.title("Outliers in Duration")

plt.show()


# =========================
# Section 5 : Key Insights
# =========================

# 1) Waiting time and eruption duration show a positive relationship.

# 2) Data has clusters.

# 3) Waiting time distribution is not uniform.


# =========================
# Section 6 : Narrative
# =========================

# The geyser dataset was analyzed using exploratory data analysis techniques.
# After cleaning and removing missing values, statistical analysis and
# visualizations revealed relationships between waiting time and eruption duration.

# The data shows clustering behavior, indicating natural grouping of eruptions.

# Overall, the dataset helps understand patterns in geyser eruption behavior.
```
