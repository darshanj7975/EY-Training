```python
import seaborn as sns
import pandas as pd

# Load Titanic dataset
df = sns.load_dataset("titanic")

# Basic information
print(df["pclass"].value_counts())
print(df["embarked"].nunique())
print(df.isnull().sum())
print(df.shape)

# Drop rows with any missing values
df_dropped = df.dropna()
print(df_dropped.shape)

# Drop rows where only 'age' or 'embarked' are missing
df_dropped2 = df.dropna(subset=['age', 'embarked'])
print(df_dropped2.shape)

# Filling missing values
# Strategy 3: Fill missing values

# Never mutate original data while exploring
df2 = df.copy()

# Fill missing age values with mean
df2['age'] = df2['age'].fillna(df2['age'].mean())

# Fill missing embarked values with mode
df2['embarked'] = df2['embarked'].fillna(df2['embarked'].mode()[0])

# Drop 'deck' column because it has too many missing values
df2 = df2.drop(columns=['deck'])

print("Missing after cleaning")
print(df2.isnull().sum())

# df2 = cleaned data

# Average fare paid by passengers from each embarkation port
result = df.groupby("embarked")["fare"].mean()

print(result)
```
